Autism spectrum disorder (ASD) is a neurodevelopmental disorder that affects how people interact with others, communicate, learn, and behave. Functional magnetic resonance imaging (fMRI) is a noninvasive and reliable tool for revealing brain function. It is well known that spontaneous fluctuations in the blood oxygen level-dependent (BOLD) signal, as measured by fMRI, are a valuable source of data for mapping human functional neural structure. fMRI, including rs-fMRI and task-based fMRI, has become one of the primary tools of cognitive neuroscience. Studies have shown that the intrinsic connectivity of the healthy brain is altered in the autistic brain, with reports of hyperconnectivity, hypoconnectivity, and/or a combination of both.
